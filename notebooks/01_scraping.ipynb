```python
import json

with open("../data/raw/discovered_comedy_urls.json", "r") as f:
    urls = json.load(f)

def is_valid_transcript_url(url):
    if url.endswith("/comedy"):
        return False
    
    if "/comedy/page/" in url:
        return False
    
    if "transcript" not in url.lower():
        return False
    
    return True

filtered_urls = [u for u in urls if is_valid_transcript_url(u)]

registry = [
    {
        "url": url,
        "comedian": None,
        "special": None
    }
    for url in filtered_urls
]
len(registry)

with open("../data/raw/comedy_registry.json", "w", encoding="utf-8") as f:
    json.dump(registry, f, indent=2)
```
