```python
import json

with open("../data/interim/comedy_preprocessed.json", "r", encoding="utf-8") as f:
    data = json.load(f)

texts = [item["text_lower"] for item in data]
corpus = " ".join(texts)
```

### Building a simple bigram model

```python
from collections import defaultdict
import random

def build_bigram_model(tokens):
    model = defaultdict(list)
    for w1, w2 in zip(tokens[:-1], tokens[1:]):
        model[w1].append(w2)

    return model

tokens = corpus.split()
bigram_model = build_bigram_model(tokens)
```

```python
def generate_text(model, start_word, length=100):
    word = start_word
    output = [word]

    for _ in range(length):
        next_words = model.get(word)
        if not next_words:
            break
        word = random.choice(next_words)
        output.append(word)
    return " ".join(output)

print(generate_text(bigram_model, start_word = "you", length=80))
```

- The generated text reproduces surface level linguistic patterns such as repetition and informal phrasing but lacks timing structure and surprise, which are the core elements of humor
